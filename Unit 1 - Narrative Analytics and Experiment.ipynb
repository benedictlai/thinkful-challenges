```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

google = pd.read_csv('googleplaystore.csv')
google.head()
```

```python
google_no_missing = google.dropna()
google_no_missing
```

```python
google['4.0+ Rating?'] = np.where(google['Rating']>=4.0, "Yes", "No")
google
```

```python
google['10,000+ Reviews?'] = np.where(google['Reviews']>=10000, "Yes", "No")
google
```

# Hypothesis

Google may save money by eliminating apps that not only score below a 4.0/5 rating, but also those apps that have a low
reviews. However, there are solutions that I will list later on to prevent them from happening if developers and companies react to feedback well.


# Exploratory Data Analysis 

```python
sns.barplot(data = google, x = '10,000+ Reviews?', y = 'Reviews')
plt.title('Reviews with 10,000+ Reviews vs. sub-10,000 reviews')
plt.subplots_adjust(top=3.00, bottom=1.37, right=1.47)
plt.show()
```

The average number of reviews that were listed as "No" (10,000+ Reviews?) were very small (1,100 an estimate) compared to the "Yes" responses for 10,000+ Reviews (averaged around 110,000 reviews for those respondents).

```python
sns.barplot(data = google, x = '4.0+ Rating?', y= 'Rating')
plt.title('Apps with a 4.0+ Rating vs. < 4.0 Rating')
plt.subplots_adjust(top=1.88, bottom=1.17, right=1.47)
plt.ylim([0, 4.5])
plt.show()
```

The average number of "4.0+ Rating?+" that were listed as "No" was 3.4 while the "Yes" responses for 4.0+ Reviews averaged about a 4.5.

```python
ttable = google[['4.0+ Rating?','Rating']].groupby('4.0+ Rating?').agg('count').reset_index()
print(ttable)
```

```python
table = google[['10,000+ Reviews?','Reviews']].groupby('10,000+ Reviews?').agg('count').reset_index()
print(table)
```

The apps that earned above a 4.0+ rating represented 78.66% while the apps that earned below a 4.0 represented 21.34%. As for Those that reached 10,000 or more reviews, 60.74% represented that, while those that had 9,999 or less reviews for that app, 39.26% represented that. The apps had a “NaN” for either variable was EXCLUDED in this experiment.

# Rollout Plan

For the low number of reviews, I plan to reach out to those people and ask them to write a review to rate this product. Having many reviews (regardless of the ratings) will bring more awareness and more revenue for the app. The way to do this is to track their phone (if they use that app) and reach out to that customer to find out what rating they get on a scale from 0 to 5. This type of reaching out to get customers to give them ratings relates to A/B testing (size of the people I survey).

As for the apps with below a 4.0 rating, we will reach out to the ones in charge of developing the app to see what improvements they can do to make the app better. Any customer who just rates the rating without the review, we need to track their phone and ask why they gave it a lower than expected rating. When the customer gives feedback on what improvements the developers and company for that app need to do, the ratings may improve assuming the parties involved with that app will listen.

It should take about 1-3 months to have this experiment of getting the developers to listen to customer's feedback to improve the average rating on the app and get the data from the customer to get the reviews to generate more revenue.

By 4-8 months after this experiment, we hope to have the results.
