```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

movies = pd.read_csv('moviedata.csv')
movies.head()

title = movies.groupby('Title').mean()

title = title.sort_values(by = 'Rating',ascending = False)[0:10]

title['Rating'].plot(kind='bar')

plt.xlabel('Title')
plt.ylabel('Rating')
plt.title('Best Rated Movies from 2006-2016 by IMDB Users')

plt.show()

print("Regardless how much those movies made and/or what the critics thought of them, these are the most well-liked films among IMDB users.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

movies = pd.read_csv('moviedata.csv')
movies.head()

title = movies.groupby('Title').mean()

title = title.sort_values(by = 'Rating',ascending = True)[0:10]

title['Rating'].plot(kind='bar')

plt.xlabel('Title')
plt.ylabel('Rating')
plt.title('Worst Movies from 2006-2016')

# show the plot
plt.show()

print("These are the worst movies by IMDB users regardless what the critics thought about the movie from 2006-2016. Only one movie had rated lower than 2/10 among users.")
```

```python
sns.jointplot(x ="Metascore", y ="Rating", data = movies)

print("In 2014, Metascore had one of the best scores for a top movie. 2016 had one of the worst movies ever among critics. 2006 had the highest score for a low quality movie among critics. </n> Despite three critics giving high marks to three of those movies, the IMDB ratings among users averaged a 5/10, which translates that the audiences thought those movies were overrated.")
```

```python
sns.stripplot(x = "Year", y = "Metascore", data = movies , jitter = True);
print("In 2014, Metascore had one of the best scores for a top movie. 2016 had one of the worst movies ever among critics. 2006 had the highest score for a low quality movie among critics.")
```

```python
sns.stripplot(x = "Year", y = "Rating", data = movies , jitter = True);
print("While 2008 had one of the best movies ever based on ratings, 2008 also had one of the worst movies ever. In 2013, one of the best movies in ratings was the lowest.")
```

```python
sns.boxplot(x  =  "Year",  y  =  "Runtime" , data = movies);
print("In 2012, most of the movies had long runtimes while 2016 had the least amount of runtimes on average.")
```
