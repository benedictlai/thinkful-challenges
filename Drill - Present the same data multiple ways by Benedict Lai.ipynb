```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movies = pd.read_csv('moviedata-IMDB.csv')
movies.head()
```

```python

movies["FirstGenre"] = movies["Genre"]
movies.head()

movies["FirstGenre"]= movies["FirstGenre"].apply(lambda x: x.split(",")[0])
movies.head()
```

```python
plt.hist(movies['Rating'], bins=50, color='green', alpha=.5)
plt.xlabel('Rating')
plt.title('Histogram for Audience Ratings')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.catplot(y='Rating', data=movies, height=5.5, kind="point", palette="pastel", ci=94, dodge=True)
sns.despine(left=True)
plt.title('Pointplot of Audience Ratings')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(movies['Rating'])
plt.title('Boxplot of Audience Ratings')
sns.despine(offset=13)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(y="Rating", data=movies, palette="Dark2_r")
plt.title('Barplot of Audience Ratings')
sns.despine(offset=35)
plt.show()
```

Pros and Cons for the four plots above:
1. Histogram:

Pros: The graph does an effective job in showing the normal distribution for the different ratings.

Cons: It fails to show the mean and the value for all of the samples are ambiguous.

2. Pointplot:

Pros: It does show the mean.

Cons: Unforunately, I was not able to find the distrbution, minimum, maximum, and standard deviation for the data.

3. Boxplot:

Pros: I was able to see the acidity volatility for the data for median, minimum, and maximum.

Cons: Unfortunately, there was more isolated data than I expected in the minimum area.

4. Barplot:

Pros: The graph showed the volatile activity for the maximum accurately.

Cons: Similar to pointplot, it fails to show a normal distrubution, minimum, maximum, and standard deviation for the data.

```python
w = sns.lmplot(y='Votes', x='Revenue (Millions)', data=movies, fit_reg=False, scatter_kws={'alpha':0.5})
w.set_ylabels("Votes")
w.set_xlabels("Revenue (Millions)")
plt.title('Scatterplot - Votes vs. Revenue (Millions)')
plt.show()
```

```python
w = sns.jointplot(y='Votes', x='Revenue (Millions)', data=movies, kind="kde", height=5, space=0)
plt.title('Votes vs. Revenue(Millions)  - Joint Kernel Desntity')
plt.show()
```

```python
w = sns.lmplot(y='Votes', x='Revenue (Millions)', data=movies, fit_reg=True, scatter_kws={'alpha':0.7})
w.set_ylabels("Votes")
w.set_xlabels("Revenue (Millions)")
plt.title('Regression Line (Scatterplot) - Votes vs. Revenue(Millions)')
plt.show()
```

Pros and Cons for the 3 plots above:

1. Scatterplot:

Pros: It showed the maxmimum and minimum for the votes and revenue (millions).

Cons: The relationship between those variables were ambiguous.

2. Joint Kernel Density Estimate:

I was not able to find Pros, which leads me to:

Cons: The graph showed a lack of mean and it failed to provide a central estimate when the variables correlate together.

3. Scatterplot (regression):

Pros: The graph did an effective job in showing a type of trend (Regardless if it was positive or not)

Cons: For future values, I need to see accurately plot lines.

```python
sns.set(style='whitegrid', color_codes=True)
sns.swarmplot(x='FirstGenre', y='Rating', data=movies);
plt.title('Swarmplot')
plt.subplots_adjust(top=3.13, bottom=2.32, right=2.55)
plt.show()
```

```python
g=sns.catplot(x='FirstGenre', y='Rating', hue='FirstGenre', data=movies, kind='bar', palette='CMRmap', ci=99)
plt.title('Barplot')
plt.subplots_adjust(top=3.13, bottom=2.32, right=2.55)
plt.show()
```

```python
sns.set(style='darkgrid', color_codes=True)
sns.stripplot(x='FirstGenre', y='Rating', data=movies);
plt.title('Stripplot - Revenue (Millions)')
plt.subplots_adjust(top=3.13, bottom=2.32, right=2.55)
plt.show()
```

```python
sns.set(style="darkgrid")
g = sns.catplot(x='FirstGenre', y='Rating', hue='FirstGenre', data=movies, kind='violin', ci=98, dodge=True, join=False)
plt.title('Violin Pointplot - FirstGenre and their Ratings')
plt.subplots_adjust(top=3.13, bottom=0.87, right=2.55)
plt.show()
```

```python
sns.boxplot(x='FirstGenre', y='Rating', hue='FirstGenre', data=movies)
plt.title('Boxplot - FirstGenre and their Ratings')
plt.subplots_adjust(top=3.13, bottom=2.32, right=2.55)
sns.despine(offset=8, trim=True)
plt.show()
```

```python
plt.hist(movies['Rating'], bins=15, color='green',  alpha=.4, label='Green')
plt.xlabel('Total Audience Ratings')
plt.title('Histogram - Audience (Ratings)')
plt.subplots_adjust(top=3.13, bottom=2.32, right=2.55)
plt.show()
```
